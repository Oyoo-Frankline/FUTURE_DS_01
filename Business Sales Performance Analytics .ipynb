```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Administrator\Desktop\FI\online_retail.csv")
df.head(20)
```

```python
df.info()
```

```python
# Removing the missing products name
df = df.dropna(subset = ["Description"])
print(df)
```

```python
#Removing the missing customner ID's
df = df.dropna(subset=["CustomerID"])
print(df)
```

```python
#Removing returns with negatuve qunatity

df = df[df["Quantity"]>0]
df.head(50)
```

```python
#Removing negative prices;

df = df[df["UnitPrice"]>0]
df.head(10)
```

```python
#Converting the date

# df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
#Creating a Revenue column
df["Revenue"] = df["UnitPrice"] * df["Quantity"]
df
```

```python
#Total Revenue
Total_Revenue = df["Revenue"].sum()
Total_Revenue
```

```python
#Total revenue is 8911407.904
```

```python
#Total Orders
Total_Orders = df["InvoiceNo"].nunique()
Total_Orders
```

```python
#Total orders = 18532
```

```python
#Total Customers
Total_customers = df["CustomerID"].nunique()
Total_customers
```

```python
#Total customers are 4338
```

```python
# Revenue Trend analysis

df["Month"] = df["InvoiceDate"].dt.to_period("M")

#Monthly report

Monthly_sales = df.groupby("Month")["Revenue"].sum()
Monthly_sales
```

```python
# visualizing
import matplotlib.pyplot as plt

Monthly_sales.plot(figsize=(13, 6), color= "red")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
#Results found

# a) The revenue increased but dropped on the month of December
# b) February has the lowest revenue
# c) November has the highest revenue

```

```python
#Top Selling Products

top_products = (
    df.groupby("Description")["Revenue"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    
)

top_products
```

```python
#Visualizing 
top_products.plot(kind="bar",figsize=(13,6), color = "red")
plt.title("Top 10 Products by Revenue")
plt.show()
```

```python
# Findings

# Paper craft, little bridge has the highest revenue


```

```python
# Best Selling Products by Quantity

top_quantity = (

    df.groupby("Description")["Quantity"]
     .sum()
     .sort_values(ascending= False)
     .head(10)
)

top_quantity
```

```python
# High value categories

def category(product):

    product = str(product)

    if "HEART" in product:
        return "Home Decor"

    elif "CAKE" in product:
        return "Kitchen"

    elif "PARTY" in product:
        return "Party Supplies"

    else:
        return "Other"

df['Category'] = df['Description'].apply(category)

```

```python
#Revenue by category

category_sales = (
    df.groupby('Category')['Revenue']
      .sum()
      .sort_values(ascending=False)
)
category_sales 
```

```python
#visualizing

category_sales.plot(kind = "bar", figsize=(13,6), color = "green")
plt.title("Category Sales Plot")
plt.show()
```

```python
#Regional performance
country_sales = (
    df.groupby("Country")["Revenue"]
    .sum()
    .sort_values(ascending = False)


)

country_sales.head(10)
```

```python
#visualizing

country_sales.head(10).plot (figsize=(13,6), kind = "bar", color = "purple")
plt.title("Top Country by Revenue")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()


```

```python
#Top customers

top_customers = (

    df.groupby("CustomerID")["Revenue"]
    .sum()
    .sort_values(ascending = False)


    
)

top_customers.head(10)
```

```python
#visualizing

top_customers.head(10).plot(figsize=(13,10), kind= "pie", color = "blue" )
plt.title("Top Customers")
plt.xlabel("Customer ID")
plt.ylabel("Revenue")
plt.show()
```
